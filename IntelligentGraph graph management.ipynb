**Graph Management with IntelligentGraph**

To enable the IntelligentGraph capabilities within a IJava notebook import the JAR, normally named and identified as follows:

```java
%jars /usr/share/intelligentgraph-0.1.0.jar
```

You will also need to 'import' the classes that you will use in your code:


```java
import pathCalc.Thing;
import pathQLModel.Resource;
import pathCalc.Evaluator;
import pathQLRepository.*;
```

You need to access an existing IntelligentGraph repository, and in this example assign it to the variable 'source'

```java
PathQLRepository source = PathQLRepository.create("http://host.docker.internal:8080/rdf4j-server","calc2graph");
```

In this context Graphs are like your own spreadsheet workbook. When you addGraph() a new Graph for your own use, any subsequent actions will take place on that grpah. Additionally the contents of that graph will remain private to you

```java
Graph graph = source.addGraph("<http://inova8.com/calc2graph/myGraph>");
```

If you have access to the Repository Workbench, yoiu will see the graph has been added to the list of contexts:<p/>


By default, any additions etc. will be made to this graph/context until you either close or remove the graph.

```java
source.closeGraph("<http://inova8.com/calc2graph/myGraph>");
```

```java
source.removeGraph("<http://inova8.com/calc2graph/myGraph>");
```
